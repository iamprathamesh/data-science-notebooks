```python
import pandas as pd
columnNames = ["RND","ADMIN","MKT","STATE","PROFIT"]
dataframe = pd.read_csv("assets/50_Startups.csv", header=None, names=columnNames)
dataframe.corr()
```

```python
X = dataframe[["RND"]]
Y = dataframe[["PROFIT"]]

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=23)

import statsmodels.api as sm
X2 = sm.add_constant(xtrain)
est = sm.OLS(ytrain, X2)
model = est.fit()
model.summary()
```

```python
X = dataframe[["RND","MKT"]]
Y = dataframe[["PROFIT"]]

xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=23)

from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(xtrain,ytrain)
pred = model.predict(xtest)

from sklearn.metrics import mean_absolute_error
MAE_SLR = mean_absolute_error(ytest,pred)
MAE_SLR
```
