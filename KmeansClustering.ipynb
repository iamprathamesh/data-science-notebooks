```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
df = pd.read_csv("assets/Cars93.csv")
le = LabelEncoder()
df2 = df.apply(le.fit_transform)
df2.head()
```

```python
X = df2[["MPG.highway","Price"]]
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4)
model = km.fit(X)
df2["clustering_vector"] = model.labels_
```

```python
from sklearn.cluster import KMeans
WCSS = []
KValue = range(2,20,1)

for k in range(2,20,1):
    km = KMeans(n_clusters=k)
    model = km.fit(X)
    WCSS.append(model.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(KValue,WCSS,'bx-',c="red")
plt.xlabel("K number of clusters")
plt.ylabel("Within cluster sum of squares")
plt.xticks(range(1,20,1))
plt.title("ELBOW CURVE")
```

```python
X = df2[["MPG.highway","Price"]]
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4)
model = km.fit(X)
df2["clustering_vector"] = model.labels_
```

```python
colors = {0: "red", 1: "blue", 2: "green", 3: "yellow"}
coloring = []
for vector in df2.clustering_vector:
    coloring.append(colors[vector])
```

```python
import matplotlib.pyplot as plt2
plt2.scatter(df2.Price,df2['MPG.highway'],c=coloring)
```
