```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder 
df = pd.read_csv("assets/Cars93.csv")
le = LabelEncoder()
df2 = df.apply(le.fit_transform)
df2.head()
```

```python
X = df2[["Price","MPG.city","MPG.highway","Passengers","Length"]]
Y = df2[["AirBags"]]
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size = 0.3, random_state = 21)
knn = KNeighborsClassifier()
params = { "n_neighbors" : range(2,30,1)}
cv = GridSearchCV(knn,params,scoring="accuracy",cv=4)
```

```python
cvmodel = cv.fit(xtrain,ytrain)
cv.best_params_
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
model = knn.fit(xtrain,ytrain)
pred = model.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest,pred)
```
