```python
import pandas as pd
columnNames = ["RND","ADMIN","MKT","STATE","PROFIT"]
df = pd.read_csv('/Users/prathameshmadur/Documents/Data_Science_notebooks/assets/50_Startups.csv', header=None, names=columnNames)
df.corr()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.RND, df.PROFIT)
```

```python
import seaborn as sb
sb.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
model1X = df[["RND"]]
model1Y = df[["PROFIT"]]
xtrainm1,xtestm1,ytrainm1,ytestm1 = train_test_split(model1X,model1Y,test_size = 0.2)
```

```python
model2X = df[["RND", "MKT"]]
model2Y = df[["PROFIT"]]
xtrainm2,xtestm2,ytrainm2,ytestm2 = train_test_split(model2X,model2Y,test_size = 0.2)
```

```python
model3X = df[["RND", "MKT", "ADMIN"]]
model3Y = df[["PROFIT"]]
xtrainm3,xtestm3,ytrainm3,ytestm3 = train_test_split(model3X,model3Y,test_size = 0.2)
```

```python
sb.distplot(df.ADMIN)
```

```python
sb.distplot(xtrainm3.ADMIN)
```

```python
from sklearn.linear_model import LinearRegression
lm1 = LinearRegression()
lm2 = LinearRegression()
lm3 = LinearRegression()
model1 = lm1.fit(xtrainm1, ytrainm1)
model2 = lm2.fit(xtrainm2, ytrainm2)
model3 = lm3.fit(xtrainm3, ytrainm3)
```

```python
predm1 = model1.predict(xtestm1)
predm2 = model2.predict(xtestm2)
predm3 = model3.predict(xtestm3)
```

```python
ytestm1["predicted_vals"] = predm1
ytestm2["predicted_vals"] = predm2
ytestm3["predicted_vals"] = predm3
```

```python
ytestm1["Residuals"] = ytestm1.PROFIT-ytestm1.predicted_vals
ytestm2["Residuals"] = ytestm2.PROFIT-ytestm2.predicted_vals
ytestm3["Residuals"] = ytestm3.PROFIT-ytestm3.predicted_vals
```

```python
from sklearn.metrics import mean_absolute_error
ytestm1
MAE = mean_absolute_error(ytestm1.PROFIT,ytestm1.predicted_vals)
print("Mean absolute error in model is %.2f"%MAE)
```

```python
ytestm2
MAE = mean_absolute_error(ytestm2.PROFIT,ytestm2.predicted_vals)
print("Mean absolute error in model is %.2f"%MAE)
```

```python
ytestm3
MAE = mean_absolute_error(ytestm3.PROFIT,ytestm3.predicted_vals)
print("Mean absolute error in model is %.2f"%MAE)
```
